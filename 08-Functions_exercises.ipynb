def final_velocity(initial_velocity, acceleration, time):
    """
    Calculate the final velocity of an object undergoing constant acceleration.

    Parameters:
    - initial_velocity: Initial velocity of the object (in meters per second)
    - acceleration: Acceleration of the object (in meters per second squared)
    - time: Time duration of acceleration (in seconds)

    Returns:
    Final velocity of the object (in meters per second)
    """
    return initial_velocity + acceleration * time

# Example usage:
initial_velocity = 10.0  # Initial velocity in meters per second
acceleration = 2.0  # Acceleration in meters per second squared
time = 5.0  # Time duration in seconds

result_final_velocity = final_velocity(initial_velocity, acceleration, time)
print(f"Final Velocity: {result_final_velocity} meters per second")

### 8.5 Απλός αρμονικός ταλαντωτής

Γράψτε μια συνάρτηση που θα υπολογίζει την θέση ενός αντικειμένου υπό την επίδραση ενός απλού αρμονικού ταλαντωτή. Δηλαδή να λύνει την εξίσωση:

$$ x(t) = A \times cos(\omega t + \phi) $$

όπου 

$ x(t) $: η θέση <br>
$ A $: το πλάτος της ταλάντωσης (cm) <br> 
$ \omega $: γωνιακή συχνότητα (radians) <br>
$ t $: χρόνος (s) <br>
$ \phi $: διαφορά φάσης (radians) <br>

Βρείτε την θέση του αντικειμένου για μια σειρά από χρόνους. 

Χρησιμοποιήστε κάποιες προκαθορισμένες τιμές (default) για το πλάτος, την γωνιακή συχνότητα, και την διαφορά φάσης σε περίπτωση που δεν δίνονται.

import math as m

def position(time, amplitude=10, angular_freq= 2 , phase_angle= m.pi/4):
    """
    Calculate the position of an object undergoing simple harmonic motion.

    Parameters:
    - time: Time, in seconds
    - amplitude: Amplitude of the motionm in cm (default value = 10)
    - angular_frequency: Angular frequency, in radians per second (default value = 2))
    - phase_angle: Phase angle, in radians (default value = $\pi/4$))

    Returns:
    Position of the object at the given time
    """
    return amplitude * m.cos(angular_freq * time + phase_angle)

# example with default values 
print('1. default values')
for i in range(10):
    print(f'- Position at time={i}s : {position(i)}')
print('='*50)    
    
# examples with modified values:
print('2. changing amplitude')
for i in range(10):
    print(f'- Position at time={i}s : {position(i, amplitude=35 )}')
print('='*50)    

print('3. changing all')
for i in range(10):
    print(f'- Position at time={i
    }s : {position(i, amplitude=35, angular_freq= 65 , phase_angle= 4  )}')

import random

num_objects = 80
data = [ random.randint(1, 100) for i in range(num_objects) ]
labels = [ random.choice( ['A', 'B'] ) for i in range(num_objects) ]

def sample_split( sample_data, sample_labels, train=70, valid=20, test=10):
    """ Split a sample (containing data and labels) in train/validation/test 
    subsamples.
    The function DOES not check if the input fractions sum up to 100! 
    user should be cautious!

    Parameters:
    sample_data      : the values of the sample to be split
    sample_labels    : the labels of the sample to be split
    train/valid/test : fractions of subsample (if given).

    Returns:
    Three tuples corresponding to each of the train, validation, 
    and test subsamples.
    Each tuple contains the data and labels of the corresponding subsample.
    """
#    print(train, test, valid)
    nobjects = len(sample_data)  # sample_data or sample_labels have the same lenght
    # find the corresponding number of objects/indeces for each subsample
    n_train = int(nobjects*train/100)
    n_valid = int(nobjects*valid/100)
    n_test = int(nobjects*test/100)
    # check for fractions
#    print(n_train, n_valid, n_test)
    
    # slice sample_data/labels accordingly
    train_data   = sample_data[:n_train]
    train_labels = sample_labels[:n_train]

    valid_data   = sample_data[n_train:n_train+n_valid]
    valid_labels = sample_labels[n_train:n_train+n_valid]

    test_data    = sample_data[n_train+n_valid:]
    test_labels  = sample_labels[n_train+n_valid:]
    
    # check to ensure that we split correctly
#    print(n_train, n_train+n_valid, n_train+n_test+n_valid)

    # NOTE: this return is just an option, you could return the 6 lists directly, 
    # or any other combination, but it should be stated crearly in the 
    # docstring for the user to understand the output
    return (train_data, train_labels), (valid_data, valid_labels), (test_data, test_labels)

def char_count(text):
    """
    Counts the frequency of each char in the given text.
    """
    chars = text.lower() #.split()
#    print(chars)
    freq = {}
    for c in chars:
        freq[c] = freq.get(c, 0) + 1
    return freq

text = input('Give me any text: ')

for key, value in char_count(text).items():
    print(f'Char: {key} -- {value} times')
